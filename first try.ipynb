```python
def encode(data, type, number ) :
    #keep only the one we need to do one hot encoding
    #X = df[['DayOfWeek','StoreType','Assortment', "Promo2SinceWeek","Promo2SinceYear" ]]
    X = data[[type]]
    # encode labels with value between 0 and n_classes-1.
    
    le = preprocessing.LabelEncoder()
    
    X_2 = X.apply(le.fit_transform)
    
    enc = preprocessing.OneHotEncoder(n_values=number)
    #fit the model
    enc.fit(X_2)
    
    # do the one hot encoder 
    onehotlabels = enc.transform(X_2).toarray()
    
    return onehotlabels
```
